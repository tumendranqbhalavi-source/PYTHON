# ***1.What is the difference between a function and a method in Python***

Ans= Function

1.A function in Python is a block of reusable code that performs a specific task.

2.It is defined using the def keyword.

3.A function can exist independently (outside of a class).

Example:

```python
def add(a, b):
    return a + b

print(add(2, 3))  # Output: 5
```

**Method**
1.A method is also a function, but it is associated with an object (class instance).

2.It is defined inside a class.

3.It usually takes self as the first parameter, which refers to the current object.

Example:

```python
class Calculator:
    def add(self, a, b):   # method
        return a + b

calc = Calculator()
print(calc.add(2, 3))  # Output: 5
```

# ***6. Explain the concept of generators in Python and how they are defined.***

Ans=**Generators in PythonConcept**

A generator is a special type of function in Python that produces a sequence of values lazily (one at a time, on demand).

Instead of return, it uses the yield keyword.

Generators automatically create iterators.

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

gen = count_up_to(5)

print(next(gen))
print(next(gen))
print(next(gen))
```

# ***7. What are the advantages of using generators over regular functions?***



## ***Ans = Advantages of Generators over Regular Functions***
## 1. Memory Efficient

Generator values produce on demand (lazy evaluation).

Normal function returns full list → takes more memory.

Example:

```python

def squares_list(n):
    return [i*i for i in range(n)]
def squares_gen(n):
    for i in range(n):
        yield i*i

print(squares_list(5))
print(list(squares_gen(5)))
```

**2. Faster Execution for Large Data**

Generators don’t wait to compute the whole list.

They give next value immediately → faster in streaming data.

**3. Cleaner Code**

Avoids using extra data structures like lists.

Code is shorter & readable with yield.

# ***10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?***

Ans= 1. **map()**

Purpose: Apply a function to every element of an iterable.

Returns: A new iterator with transformed elements.

 Example:

```python
nums = [1, 2, 3, 4]

result = map(lambda x: x*2, nums)
print(list(result))   # [2, 4, 6, 8]
```

2. **filter()**

Purpose: Filter out elements from an iterable based on a condition (True/False).

Returns: Only the elements that satisfy the condition.

Example:

```python
nums = [1, 2, 3, 4, 5, 6]

result = filter(lambda x: x % 2 == 0, nums)
print(list(result))   # [2, 4, 6]
```

3. **reduce()**

Purpose: Reduce all elements of an iterable into a single value (using cumulative function). Returns: One final result.

Note: reduce() is in the functools module, so you must import it.

Example:

```python
from functools import reduce

nums = [1, 2, 3, 4]

result = reduce(lambda x, y: x + y, nums)
print(result)
```

# ***1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.***

```python
def sum_even(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

nums = [1, 2, 3, 4, 5, 6]
print("Sum of even numbers:", sum_even(nums))
```

```python
def square_list(numbers):
    result = []
    for num in numbers:
        result.append(num ** 2)
    return result

nums = [1, 2, 3, 4, 5]
print("Squares:", square_list(nums))
```

# ***4. Write a Python function that checks if a given number is prime or not from 1 to 200.***

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

for num in range(1, 201):
    if is_prime(num):
        print(num, "is Prime")
    else:
        print(num, "is Not Prime")
```

# ***5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.***

```python
def fibonacci(n_terms):
    a, b = 0, 1
    for _ in range(n_terms):
        yield a
        a, b = b, a + b

for num in fibonacci(10):
    print(num)
```

# ***6. Write a generator function in Python that yields the powers of 2 up to a given exponent.***

```python
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

for val in powers_of_two(5):
    print(val)
```

# ***7. Implement a generator function that reads a file line by line and yields each line as a string.***




```python

with open("sample.txt", "w") as f:
    f.write("This is line 1.\n")
    f.write("This is line 2.\n")
    f.write("This is line 3.")

def read_file_line_by_line(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()

for line in read_file_line_by_line("sample.txt"):
    print(line)
```

# ***9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.***

```python
c_to_f = lambda c: (c * 9/5) + 32

celsius = [0, 20, 37, 100]

fahrenheit = list(map(c_to_f, celsius))

print("Celsius:    ", celsius)
print("Fahrenheit: ", fahrenheit)
```

# ***10. Create a Python program that uses `filter()` to remove all the vowels from a given string.***

```python
is_consonant = lambda ch: ch.lower() not in "aeiou"

def remove_vowels(text):
    return "".join(filter(is_consonant, text))
s = "Hello World"
print("Original:", s)
print("Without vowels:", remove_vowels(s))
```
