2. How does broadcasting work in NumPy

Ans= Broadcasting in NumPy is a way of performing arithmetic operations on arrays of different shapes without explicitly copying data. When two arrays don‚Äôt have the same shape, NumPy automatically **‚Äústretches‚Äù the smaller array** along the missing dimensions so they match, and then applies the operation element-wise. For example, if you add a 1D array to a 2D array, the 1D array is broadcast across each row or column as needed. This makes code shorter, faster, and more memory-efficient since no extra data is created.

# ***4. Explain the use of the groupby() method in Pandas***

Ans= The groupby() method in Pandas is used to split data into groups based on the values of one or more columns, and then perform operations like aggregation, transformation, or filtering on each group separately. It follows a simple process often called ‚ÄúSplit-Apply-Combine‚Äù:

Split ‚Äì The DataFrame is divided into groups based on a column(s).

Apply ‚Äì A function (like sum(), mean(), count(), etc.) is applied to each group.

Combine ‚Äì The results are merged back into a new DataFrame or Series.

üëâ Example: If you have sales data with a "Region" column, using groupby("Region").sum() will give total sales for each region.

# ***6. What are the differences between NumPy arrays and Python lists***

Ans= Here‚Äôs a short comparison of **NumPy arrays vs Python lists**:

1. **Data Type** ‚Äì NumPy arrays store elements of the **same data type**, while Python lists can hold elements of **different types** (e.g., int, float, string together).
2. **Memory Efficiency** ‚Äì NumPy arrays are **more compact and memory-efficient** because they use fixed-type data, whereas lists are less efficient.
3. **Speed** ‚Äì NumPy arrays are **much faster** for mathematical and numerical operations since they use optimized C/C++ code in the background.
4. **Functionality** ‚Äì NumPy provides a wide range of **vectorized operations** (like element-wise addition, multiplication, broadcasting), which are not directly available with lists.
5. **Dimensionality** ‚Äì NumPy supports **multi-dimensional arrays (matrices, tensors)**, while Python lists are mostly 1D (though you can nest them, but it‚Äôs less efficient).

# ***17.  Explain the difference between apply() and map() in Pandas***

Ans= In Pandas, both **`apply()`** and **`map()`** are used to apply functions to data, but they differ in scope and flexibility. The **`map()`** function is mainly used with Pandas Series (1D data) and applies a function element-wise, which makes it simple and efficient for transforming or mapping values in a single column. On the other hand, the **`apply()`** function works with both Series and DataFrames. When used on a Series, it behaves similarly to `map()`, but when applied to a DataFrame, it can apply a function across rows or columns, giving it more flexibility. In short, `map()` is best for element-wise operations on a single column, while `apply()` is more powerful for applying custom functions across rows or columns in a DataFrame.

# ***21.  Why is NumPy‚Äôs array slicing faster than Python‚Äôs list slicing***

Ans= NumPy‚Äôs **array slicing** is faster than Python‚Äôs **list slicing** because of the way data is stored and accessed in memory. NumPy arrays are stored in **contiguous blocks of memory** with a fixed data type, which allows direct access using memory offsets without copying data. When you slice a NumPy array, it usually creates a **view** of the original data (not a full copy), so it‚Äôs very efficient.

In contrast, Python lists are collections of **pointers to objects** scattered in memory. When slicing a list, Python has to create a **new list** and copy references to each element, which makes it slower and more memory-intensive.
